```python
columns_to_keep = ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']
df = df[columns_to_keep]
df.head()
```

```python
df = df.drop_duplicates()
```

```python
print("5b. Missing values", df.isnull().sum())
```

```python
df = df.dropna(subset = ['continent'])
```

```python
df = df.fillna(0)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
```

```python
df_groupby = df.groupby('continent').max().reset_index()
print("Aggregated Data", df_groupby)
```

```python
df_groupby['total_deaths_to_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['gdp_per_capita'], kde = True , bins = 30)
plt.title('Distribution of GDP per capita')
plt.xlabel('GDP per capita')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = df , x = 'total_cases',y = 'gdp_per_capita')
plt.title('Total Cases vs GDP per capita')
plt.show()
```

```python
sns.pairplot(df_groupby)
plt.show()
```

```python
sns.catplot(data = df_groupby , x = 'continent',y ='total_cases', kind = 'bar', height = 5,aspect =2)
plt.title('Total cases by continent')
plt.show()
```

```python
df_groupby.to_csv('df_groupby.csv',index = False)
```
